```python
from qiskit.circuit import Parameter, QuantumCircuit
from qiskit_qm_provider import IQCCProvider, QMEstimatorV2, QMEstimatorOptions
from qiskit.quantum_info import SparsePauliOp
import numpy as np
from qiskit import transpile

backend = IQCCProvider().get_backend("arbel")
```

```python
qc = QuantumCircuit(1)
param = Parameter("param")
qc.rx(param, 0)
new_qc = transpile(qc, backend=backend)
obs = SparsePauliOp.from_list([("X", 1.2), ("Y", 1.3)]).apply_layout(new_qc.layout)

estimator = QMEstimatorV2(backend=backend, options=QMEstimatorOptions(input_type="INPUT_STREAM"))

job = estimator.run([(new_qc, obs, np.array([np.linspace(0, 2*np.pi, 10)]))])
```

```python
estimator._job._execution_plans[0].obs_indices
```

```python
from qm import generate_qua_script
print(generate_qua_script(job.program))
```
